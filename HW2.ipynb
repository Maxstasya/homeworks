#Задание №1
Необходимо выделить все биграммы в тексте и сохранить их список в какую-либо переменную.

```python
#Создадим массив
#Добавим элемент i списка из массива w с последующим элементом c повторонием цикла по количеству элементов массива - 1
text = 'варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг'
w = text.split() #Создадим массив слов
bigrams = [(w[i], w[i+1]) for i in range(len(w)-1)] 
print(bigrams)
```

#Задание № 2 
+ Распечатать текущую директорию и список всех файлов и папок в ней. Посчитать и распечатать количество файлов.
- Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt 

```python
import os 
print(os.getcwd()) #показать текущую директорию
os.remove('poems.txt') #удаляем образующийся в цикле файл 
for root, dirs, files in os.walk('.'): #указываем путь к текущей папке и печатаем список файлов
    print (dirs) 
    print (files) 
    print(len(files)) #количество файлов
for root, dirs, files in os.walk('.'): 
    for name in files:
         with open(os.path.join(root,name),'r',encoding='utf-8', errors = 'ignore') as f: #собираем путь к папке и читаем его
            poem = f.read() #прочитываем файл 
            with open('poems.txt','a',encoding='utf-8', errors = 'ignore') as d: #добавляем файл и записываем прочитываемые
                d.write(poem)
                print(poem)
   
```

10. Распечатать с 20 по 30 символ этой строки (включительно).
11. Разбить текст на биграммы, записать их в переменную и посчитать частотность каждой биграммы. Биграммы должны быть в виде списка кортежей.

```python
#1.Скопировать текст, сохранить его в переменную:
rhyme = """ Удивительная кошка
Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!"""
#2.Узнать длину текста в символах посредством len
print(len(rhyme))
#3.Добавить обрезку strip указанных знаков препинания для каждого элемента w в указанном массиве rhyme
print([ w.strip("!?.,-") for w in rhyme.split()])
#4.Объединить слова из получившегося списка в строку через пробел посредством join
print(' '.join([ w.strip("!?.,-") for w in rhyme.split()])) 
#5.Добавить изменение регистра на нижний элемента lower в массиве, если он написан с заглавной title
print([ w.lower() for w in rhyme.split() if w.title()])
#6.Длина в словах split указывает элементом слово 
print(len(rhyme.split()))
#7.Количество уникальных слов. Надо создать множество, элементы которого уникальны
a = set(rhyme.split())
print(len(a))
#8.Распечатать последние 10 слов с помощью среза
print(rhyme.split()[-10:])
#9.Соединить полученный в пункте 3 список в строку. Слова в строке написаны через знак переноса строки \n
print('\n'.join([ w.strip("!?.,-") for w in rhyme.split()][-10:]))
#10.Распечатать с 20 по 30 символ этой строки включительно
print(' '.join([ w.strip("!?.,-") for w in rhyme.split()][-10:])[20:31])
#11.Разбить текст на биграммы. Добавим элемент списка i из массива w с последующим элементом c повторонием цикла по количеству элементов массива -1 . Посчитать частотность каждого элемента i в списке bigrams по количесвую элементов списка
bigrams = [(rhyme.split()[i], rhyme.split()[i+1]) for i in range(len(rhyme.split())-1)]
print(bigrams)
print([bigrams.count(bigrams[i]) for i in range(len(bigrams))])
```
