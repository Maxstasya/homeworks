#Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.


```python
from pymystem3 import Mystem
def fz (ffr, ffw):#так создается функция, в которой два аргумета file for reading(и не только) и file for writing (куда все должно записаться)
    with open (ffr, 'r', encoding = 'utf-8') as f:#тут открываем файл для чтения 
        text = f.read() #читаем
        punct = ' '.join([word.strip("!«»?.,-") for word in text.split()]) #избавляемся от введенной пунктуации посредством strip и объединяем элементы списка без пунктуации в строку посредством join
        t = punct.replace("—","") #убираем тире, которое не обрезается стрипом, так как не является составной частью отдельного элемента word
        textwp = ' '.join([w for w in t.split() if w.isalnum()]) #используем условие isalnum (для букв и цифр), чтобы проверялся каждый элемент списка на соответствие и объединяем результат с join в строку, чтобы удостовериться в избавлении от пунктуации
        m = Mystem()
        lemmas = m.lemmatize(textwp) #лемматизируем текст без пунктуации
        q=''.join(lemmas) #объединяем леммы в строку
    with open (ffw, 'w', encoding = 'utf-8') as r: #открываем файл для записи 
        for i in q:
            r.write(i) #записываем леммы в файл 
        f.close
fz ('литературныеанекдоты.txt','lems.txt') #указываем аргументы функции ffr  и ffw
with open('lems.txt', 'r') as f:
        b = f.read()
        print(b) #проверяем, выводя чтение файла 
```

#Задание №2

Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
with open('lems.txt', 'r') as f: #открываем файл с леммами 
    lems = f.read()
with open ('rustopwords.txt','r', encoding='utf-8') as b: #открываем файл для чтения со стопсловами 
    stopwords = b.read().split('\n') #читая, создаем из него упорядоченный список
    wsp = ' '.join([w for w in lems.split() if w not in stopwords]) #without stopwords очищаем леммы от стопслов и объединяем в строку( для каждого элемента lems если он не в stopwords)
from collections import Counter
acount = Counter(wsp.split()) #вычисляем частотные словаря посредством импорта Counter
d = {} #создаем словарь, где ключ(key) - слово, а значение(value) -  частотность
for word in acount: #для подсчета абсолютной чистоты 
    if word in d : 
        d[word] +=1
    else:
        d[word] = 1
length = len(wsp.split()) #длина текста, очищенного от стопслов
print(length) 
for k,v in d.items(): #для подсчета относительной частоты 
    ocount = v/length * 1000000 #согласно формуле подсчета в конспекте, так что ключ - слово, частота - значение 
    d[k] = ocount
fr = sorted (d.items(), key = lambda x: x[1], reverse = True) #так на семинаре сказали делать сортировку
otnosit = fr[:20]
print(acount.most_common(20))
print(otnosit)
```
